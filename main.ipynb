```python
#importdata Moon
df = pd.read_csv("space_apps_2024_seismic_detection/data/lunar/training/data/S12_GradeA/xa.s12.00.mhz.1970-01-19HR00_evid00002.csv", sep=",", header=None, names=["Absolute Time", "Relative Time", "Velocity"], na_filter=False)
display(df)
```

```python
#Removing Relative Time because i won't be working with it
df = df.drop(columns=['Relative Time'])
#Removing the 1st row
df = df.drop([0])

display(df)
display(df.info())
```

```python
#Snapshotting the Absolute Tiem table for use later
Total_time = df['Absolute Time']
#Splitting absolute time into Hours Minutes and Seconds
Hours = df['Absolute Time'].str[11:13]
display(Hours)
Minutes = df['Absolute Time'].str[14:]
Minutes = Minutes.str[:2]
display(Minutes)
Seconds = df['Absolute Time'].str[17:]
Seconds = Seconds.str[:2]
display(Seconds)
```

```python
#Drop Absolute time for in exhange for the easier to read Hours/Minutes/Seconds
df = df.drop(columns=['Absolute Time'])
df['Seconds'] = Seconds
df['Minutes'] = Minutes
df['Hour'] = Hours
```

```python
#Change order of columns
df = df.loc[:,['Hour','Minutes','Seconds', 'Velocity']]
display(df)
```

```python
#working with the snapshotted 
Total_time = Total_time.str[11:]
Total_time = Total_time.str[:8]
display(Total_time)
```

```python
#Converting Velocity to numbers
df['Velocity'] = df['Velocity'].astype(float)
display(df)
display(df.info())
```

```python
#Displays the average, minimal value, maximum value of velocity
display(df['Velocity'].mean())
display(df['Velocity'].min())
display(df['Velocity'].max())


#This value will be used for discarding irrelevant data
median = df['Velocity'].max() + df['Velocity'].min()

display(median)
```

```python
df_high = df.drop(df[df.Velocity < median].index)
df_low = df.drop(df[df.Velocity > -median].index)
display(df_high)
display(df_low)
```

```python
#Merge the previous dataframs

#This was another path but i decided against it 
# df_filtered = df_high.merge(df_low, how='outer', on='Velocity')
# display(df_filtered)


#Outer joins the High and Low dataframes, based on their indexes
df_filtered = df_high.merge(df_low, how='outer', left_index = True, right_index = True)
display(df_filtered)
```

```python
#Replaces the Null values, then deletes the unused columns made after the join
df_filtered.Hour_x.fillna(df_filtered.Hour_y, inplace=True)
df_filtered.Minutes_x.fillna(df_filtered.Minutes_y, inplace=True)
df_filtered.Seconds_x.fillna(df_filtered.Seconds_y, inplace=True)
df_filtered.Velocity_x.fillna(df_filtered.Velocity_y, inplace=True)
del df_filtered['Hour_y']
del df_filtered['Minutes_y']
del df_filtered['Seconds_y']
del df_filtered['Velocity_y']
#Rename the DF since _x specification is not required
df_filtered = df_filtered.rename(columns={'Hour_x': 'Hour', 'Minutes_x': 'Minutes', 'Seconds_x': 'Seconds', 'Velocity_x': 'Velocity'})

display(df_filtered)
```
